```python
import pandas as pd
```

```python
df = pd.read_csv("Benign_malignant_indices.csv")
```

```python
df.head()
```

```python
def get_benign_data():
    benign_indices = df.benign.values
    malignant_indices = df.malignant.values
    
    for index in benign_indices:
        benign_image_path = f"Dataset_BUSI_with_GT/benign/benign ({index}).png"
        benign_mask_path = f"Dataset_BUSI_with_GT/benign/benign ({index})_mask.png"
        malignant_image_path = f"Dataset_BUSI_with_GT/malignant/malignant ({index}).png"
        malignant_mask_path = f"Dataset_BUSI_with_GT/malignant/malignant ({index})_mask.png"
        benign_images.append(plt.imread(benign_image_path)[:,:,0])
        benign_masks.append(plt.imread(benign_mask_path))
        malignant_images.append(plt.imread(malignant_image_path)[:,:,0])
        malignant_masks.append(plt.imread(malignant_mask_path))
    return benign_images, benign_masks, malignant_images, malignant_masks
```

```python
benign_images = []
benign_masks = []
malignant_images = []
malignant_masks = []
get_benign_data()
```
